
# OVER SAMPLING

```python
#  WE prefer OVERSAMPLING over UNDERSAMPLING...as there's a possibility
#  of loosing important data
```

```python
from imblearn.combine import SMOTETomek
```

```python
smk = SMOTETomek()
X_ovr,Y_ovr = smk.fit_resample(X,Y)
```

```python
X_ovr.shape,Y_ovr.shape
```

# Was Taking forever to run

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
os = RandomOverSampler()
```

```python
X_ovr,Y_over = os.fit_resample(X,Y)
```

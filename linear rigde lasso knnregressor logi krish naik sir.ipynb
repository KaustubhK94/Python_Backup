```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=66)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
train_score = []
test_score = []
for i in range(1,40,2):
 knn = KNeighborsRegressor(n_neighbors=i,)
 knn.fit(X_train,y_train)
 pred_i = knn.predict(X_test)
 train_score.append(knn.score(X_train,y_train))
 test_score.append(knn.score(X_test,y_test))

plt.figure(figsize=(10,6))
plt.plot(range(1,40,2),train_score,color='blue', linestyle='dashed', 
         marker='o',markerfacecolor='red', markersize=10)
plt.plot(range(1,40,2),test_score,color='blue', linestyle='dashed', 
         marker='o',markerfacecolor='yellow', markersize=10)
plt.title('R^2 vs. K Value')
plt.xlabel('K')
plt.ylabel('R^2 Score')
```

```python
Knn = KNeighborsRegressor(n_neighbors=10)
```

```python
k1 = Knn.score(X_train, y_train)
k2 = Knn.score(X_test, y_test)
print("R^2 of KNN Regressor on training set: {:.3f}".format(k1))
print("R^2 of KNN Regressor on test set: {:.3f}".format(k2))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
df = load_breast_cancer()
#Independant features
X =pd.DataFrame(df["data"],columns = df["feature_names"])
```

```python
df
```

```python
X.head()
```

```python
##Dependant Feature 
y = pd.DataFrame(df["target"],columns=["Target"])
```

```python
y
```

```python
y["Target"].value_counts()
#this is balanced dataset
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.33,random_state=42)
```

```python
 params = [{"C":[1,5,10]},{"max_iter":[100,150]}]
```

```python
model1 = LogisticRegression(C =100,max_iter=100)
```

```python
model = GridSearchCV(model1,param_grid=params,scoring="f1",cv=5)
```

```python
model.fit(X_train,y_train)
```

```python
model.best_params_
model.best_score_
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
confusion_matrix(y_pred,y_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
left = pd.DataFrame({'key': ['K0', 'K1 ', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
```

```python
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})    
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,how="inner",on="key")
```

```python
lefts = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
rights = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
lefts
```

```python
rights
```

```python
pd.merge(lefts,rights,on=["key1","key2"])
```

```python
pd.merge(lefts,rights,how = "right",on=["key1","key2"])
```

```python
pd.merge(lefts,rights,how = "left",on=["key1","key2"])
```

```python
pd.merge(lefts,rights,how = "outer",on=["key1","key2"])
```

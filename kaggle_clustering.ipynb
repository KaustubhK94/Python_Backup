```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
df_scaled = sc.fit_transform(df)
```

```python
df_scaled
```

```python
df_scaled.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca.fit(df_scaled) 
```

```python
scaled_pca = pca.transform(df_scaled)
```

```python
scaled_pca
```

```python
pca
```

```python
scaled_pca.shape
```

```python
print(pca.explained_variance_)
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.cumsum())
```

```python
from yellowbrick.cluster import (kelbow_visualizer,silhouette_visualizer)
```

```python
km = KMeans(max_iter=300, n_init=10)
elbow_k = kelbow_visualizer(km, scaled_pca, k=(2, 10))
```

```python
print("The best number of cluster is: ", elbow_k.elbow_value_)
kmeans = KMeans(n_clusters=elbow_k.elbow_value_ , n_init=10, max_iter=300)
label_pred = kmeans.fit_predict(scaled_pca)
label_pred
```

```python
len(label_pred)
```

```python
#### UNIQUE VALUES OFF THE ARRAY i.e. predicted labels....
np.unique(label_pred)
```

```python
silhouette = silhouette_visualizer(kmeans, scaled_pca)
print("The Average Silhouette score is: ", silhouette.silhouette_score_)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(scaled_pca[:,0],scaled_pca[:,1],c=label_pred,cmap="coolwarm",alpha=0.3)
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
```

```python
# make variables that indicates 1 if observations was missing else it's 0
#  np.where returns indices of elements in input array where the 
# specifeid condition is satisfied
# Understand the realtionship between missing values and sales price

data = df.copy()
for feat in feat_na:
    data[feat]=np.where(data[feat].isnull(),1,0)
    # calculate mean saleprice where information is missing or present
    data.groupby(feat)["SalePrice"].median().plot.bar()
    plt.title(feat)
    plt.show()
```

# Continous Variable

```python
continous_feat = [feat for feat in num_feat if feat not in descerte_feat+year_feat]
continous_feat
```

```python
#### Analyze the continous values by histogram to
#### Understand the distribution 
data = df.copy()
for feat in continous_feat:
    data.groupby(feat)["SalePrice"].median()
    plt.scatter(data[feat],data["SalePrice"])
    plt.xlabel(feat)
    plt.ylabel("SalePrice")
    plt.title(f"{feat} Vs SalePrice")
    plt.show()
```

```python
data = df.copy()
for feat in continous_feat:
    data[feat].hist(bins=25)
    plt.xlabel(feat)
    plt.ylabel("Count")
    plt.title(f"{feat}")
    plt.show()
```

```python
### To convert log normal distributiion into normal distribution.....
```

```python
# #### ONLY WORKS FOR LOG NORMAL DISTRIBUTION
data = df.copy()
for feat in continous_feat:
    if 0 in data[feat].unique():
        pass
    elif feat == "SalePrice":
        pass
    else:
        data[feat] = np.log(data[feat])
        plt.scatter(data[feat],data["SalePrice"])
        plt.xlabel(f"{feat}")
        plt.ylabel("Sales Price")
        plt.title(f"{feat} vs Sales Price")
        plt.show()     
```

```python
data = df.copy()
for feat in continous_feat:
    if 0 in data[feat].unique():
        pass
    else:
        data[feat] = np.log(data[feat])
        data.boxplot(column=feat)
        plt.ylabel(f"{feat}")
        plt.title(f"{feat}")
        plt.show()     
```
